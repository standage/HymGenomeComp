# Genome Composition in Hymenoptera

Interval loci (iLoci) provide a mechanism for breaking down an annotated genome sequence into manageable chunks, each representing a gene (or set of overlapping genes), an intergenic region, or an incompletely assembly fragment of the genome. This document uses iLoci to explore genome composition in 12 Hymenopteran species.

## Data inputs

Genome annotations (in GFF3 format) were collected from NCBI for 11 of the 12 species. Annotations for the 12th species (*Polistes dominula*) were produced in-house and are pending publication.

The [LocusPocus program](http://aegean.readthedocs.org/en/v0.10.2/locuspocus.html) was used to process each annotation and produce a new GFF3 file for each species containing iLocus annotations. These GFF3 files are the inputs to this procedure.

## Unannotated sequences

LocusPocus currently doesn't report iLoci for unannotated genomic sequences. To make sure we account for the entire genome sequence, we analyze all `##sequence-region` pragmas and note those with no associated annotations. This function determines the length of a sequence based on its `##sequence-region` pragma (see the [GFF3 spec](http://www.sequenceontology.org/gff3.shtml)).

```python
import re

def seqlen(pragmastring):
    """
    Determines the length of a genomic sequence from its corresponding GFF3 
    pragma record.
    """
    pattern = "##sequence-region\s+(\S+)\s+(\d+)\s+(\d+)"
    seqmatch = re.search(pattern, pragmastring)
    assert seqmatch, "unable to parse sequence region pragma: %s" % pragmastring

    seqid = seqmatch.group(1)
    start = int(seqmatch.group(2))
    end   = int(seqmatch.group(3))
    assert end >= start, "end < start: %s" % pragmastring

    seqlength = end - start + 1
    return seqid, seqlength

def test_seqlen():
    """
    Unit tests for seqlen function. Throws an assertion error in case of
    failure, so no output is good output!
    """
    test_strings = { "##sequence-region   chr1 1 1000000":      1000000,
                     "##sequence-region\tscaf0001   401   600": 200,
                     "##sequence-region   2L 1001 10000\n":     9000 }
    
    for pragmastring in test_strings:
        length = test_strings[pragmastring]
        seqid, test_length = seqlen(pragmastring)
        assert length == test_length, "seqlen test failed: %s" % pragmastring

test_seqlen()
```

## Overlapping iLoci

In general iLoci are intended to be distinct non-overlapping genomic regions. However, in gene-dense regions there will occasionally be some overlap in the genes' regulatory regions. We do not want to count these overlapping regions twice when we're giving a breakdown of genome composition, so this function is used to collapse a set of overlapping iLoci into a single gene-containing region for record-keeping purposes.

```python
from intervaltree import Interval

def merge_loci(loci):
    """
    Input is a set of interval objects, each corresponding to an iLocus.
    Output is a new interval object containing all the iLoci.
    """
    loci = list(loci)
    start, end = 0, 0
    for locus in loci:
        start = min(start, locus.begin)
        if start == 0:
            start = locus.begin
        end = max(end, locus.end)
    return Interval(start, end, loci)

def test_merge_loci():
    """
    Unit tests for merge_loci function.
    """
    testint = merge_loci((Interval(1, 3, "dude"),
                          Interval(2, 4, "sweet")))
    assert testint.begin == 1 and testint.end == 4
    
    testint = merge_loci((Interval(6, 9,  "rad"),
                          Interval(7, 30, "wicked")))
    assert testint.begin == 6 and testint.end == 30

test_merge_loci()
```

## GFF3 parsing

Here's where things get hairy.

* **Step 1**: Walk through the GFF3 file line-by-line. If the line is a `##sequence-region` pragma, store the length of the corresponding sequence. If the line is an iLocus annotation, store it in an interval tree (for efficient interval queries) and discard the length computed from the `##sequence-region` pragma. Ignore all other lines in the GFF3 file.
* **Step 2**: All sequence lengths remaining in memory correspond to unannotated sequences. Compute their sum and store.
* **Step 3**: Iterate over all iLoci in the interval trees and add their lengths to the appropriate category (gene-containing, gene-less internal, or gene-less fragment). For iLoci that overlap other iLoci, verify the assumption that they all contain genes, and merge before computing length.

```python
from intervaltree import IntervalTree
from collections import namedtuple
GenomeComposition = namedtuple("GenomeComposition",
                               "genes internal terminal unannot")
def parse_gff3(fp):
    """
    Input:  handle to GFF3 data (LocusPocus output)
    Output: a simple `GenomeComposition` data structure giving a breakdown of
            the number of nucleotides in each of 4 categories
    """
    #----------
    # Step 1
    #----------
    
    # Dictionary of interval trees, indexed by sequence ID, for storing
    # iLoci
    loci = {}
    
    # Dictionary of sequence lengths, indexed by sequence ID
    seqlengths = {}
    
    for line in fp:
        if line.startswith("##sequence-region"):
            seqid, slength = seqlen(line)
            seqlengths[seqid] = slength
        elif "\tlocus\t" in line:
            fields = line.rstrip().split("\t")
            assert len(fields) == 9
            seqid = fields[0]
            start = int(fields[3])
            end = int(fields[4]) + 1 # Python interval tree is [)
            if seqid not in loci:
                loci[seqid] = IntervalTree()
            loci[seqid][start:end] = line
            if seqid in seqlengths:
                seqlengths.pop(seqid, None)


    #----------
    # Step 2
    #----------
    unannot = 0
    for seqid in seqlengths:
        unannot += seqlengths[seqid]


    #----------
    # Step 3
    #----------
    withgenes = 0
    sansgenes = 0
    ends = 0
    merged_loci = {}
    for seqid in loci:
        merged_loci[seqid] = IntervalTree()
        for locus in loci[seqid]:
            if merged_loci[seqid][locus.begin:locus.end]:
                continue

            overlap = loci[seqid][locus.begin:locus.end]
            assert len(overlap) > 0
            if len(overlap) == 1:
                locuslength = locus.end - locus.begin
                if ";gene=" in locus.data:
                    withgenes += locuslength
                else:
                    sansgenes += locuslength
                    if ";end=true" in locus.data:
                        ends += locuslength
            else:
                cont = True
                while cont:
                    count = len(overlap)
                    for testlocus in overlap:
                        assert ";gene=" in testlocus.data
                    merged_locus = merge_loci(overlap)
                    overlap = loci[seqid][merged_locus.begin:merged_locus.end]
                    cont = len(overlap) > count
                merged_loci[seqid].add(merged_locus)
                locuslength = merged_locus.end - merged_locus.begin
                withgenes += locuslength
    
    return  GenomeComposition(withgenes, sansgenes - ends, ends, unannot)
```

## Load data into a DataFrame

Now that we've defined our GFF3 parsing procedure, let's go ahead and run it for each of the 12 species, and store the results in a data frame for analysis and visualization.

```python
import pandas as pd

# Dictionaries will end up as columns in the data frame
withgenes = {}
internal = {}
terminal = {}
unannot = {}
withgenes_perc = {}
sansgenes_perc = {}

# Process the GFF3 files (takes a good 30-60 seconds)
species_list = "Ador Aflo Amel Bimp Bter Dmel Hsal Mrot Nvit Pdom Sinv Tcas".split(" ")
for species in species_list:
    infile = "%s.iloci.gff3" % species
    with open(infile, "r") as fp:
        comp = parse_gff3(fp)
        withgenes[species] = comp.genes
        internal[species] = comp.internal
        terminal[species] = comp.terminal
        unannot[species] = comp.unannot
        total = comp.genes + comp.internal + comp.terminal + comp.unannot
        withgenes_perc[species] = float(comp.genes) / float(total)
        sansgenes_perc[species] = 1.0 - withgenes_perc[species]

# Create and print the data frame
# Hat tip: http://www.gregreda.com/2013/10/26/intro-to-pandas-data-structures/
dfdata = { "Genes": withgenes, "Internal": internal, "Terminal": terminal,
           "Unannotated": unannot, "PercGenic": withgenes_perc,
           "PercNongenic": sansgenes_perc }
compdata = pd.DataFrame(dfdata)
compdata
```

## Genic vs non-genic

Perhaps the simplest breakdown of genome composition is genic vs non-genic. The code below borrows heavily from [this page](http://nbviewer.ipython.org/github/cs109/content/blob/master/lec_03_statistical_graphs.ipynb).

```python
%matplotlib inline
import brewer2mpl
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams

def remove_border(axes=None, top=False, right=False, left=True, bottom=True):
    """
    Minimize chartjunk by stripping out unnecesasry plot borders and axis ticks
    
    The top/right/left/bottom keywords toggle whether the corresponding plot border is drawn
    """
    ax = axes or plt.gca()
    ax.spines['top'].set_visible(top)
    ax.spines['right'].set_visible(right)
    ax.spines['left'].set_visible(left)
    ax.spines['bottom'].set_visible(bottom)
    
    #turn off all ticks
    ax.yaxis.set_ticks_position('none')
    ax.xaxis.set_ticks_position('none')
    
    #now re-enable visibles
    if top:
        ax.xaxis.tick_top()
    if bottom:
        ax.xaxis.tick_bottom()
    if left:
        ax.yaxis.tick_left()
    if right:
        ax.yaxis.tick_right()

ticks = range(len(species_list))
ticks = [i+1.5 for i in ticks]
plt.bar(range(1, len(species_list)+1), compdata["PercGenic"], color="#B2182B", label="Genic")
plt.bar(range(1, len(species_list)+1), compdata["PercNongenic"], bottom=compdata["PercGenic"], color="#2166AC", label="Non-genic")
plt.xticks(ticks, species_list, rotation="horizontal")
plt.ylabel("Proportion of genome")
plt.xlabel("")
remove_border()
plt.show()
```
